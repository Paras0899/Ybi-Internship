# Q 1 - What is the total cost by aircraft type for the year?

### Total cost = Hours flown * Costs/hour

```python
finaldf.columns
```

```python
finaldf['Total cost'] = finaldf['hours flown'] * finaldf['Costs per flight hour']
```

```python
finaldf
```

```python
X = finaldf['Aircraft Type']
Y = finaldf['Total cost']
fig , ax = plt.subplots()
ax.bar(X,Y);
ax.set_title('Total cost by aircraft type')
ax.set_ylabel('Total cost')
ax.set_xlabel('Aircraft Type')
```

# Q2 - Which aircraft type  has the lowest cost per seat per km flown?

```python
finaldf.columns
```

```python
finaldf['Cost'] = finaldf['Total cost'] / (finaldf['Number of Seats'] * finaldf['hours flown'] * finaldf['Ave. Speed (km/h)'] )
```

```python
finaldf
```

```python
X1 = finaldf['Aircraft Type']
Y1 = finaldf['Cost']
fig , ax = plt.subplots()
ax.bar(X1,Y1);
ax.set_title('Cost Per seat per km')
ax.set_ylabel('cost')
ax.set_xlabel('Aircraft Type')
```

### Index of Row With Minimum Cost
